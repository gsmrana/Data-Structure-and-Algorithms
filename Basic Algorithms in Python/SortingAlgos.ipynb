```python
import random
# take the left element as pivot
# place the smallest element to left
# plece the largest element to right
# swap the pivot element to its correct position
def QuickSort_Partition2(data, left, right, reverse=False):
    pivot = data[left]
    pivot_index = left
    
    compareFunc = lambda a, b : a <= b
    if reverse: compareFunc = lambda a, b : a >= b
    
    for i in range(left + 1, right + 1):
        if compareFunc(data[i], pivot):
            pivot_index += 1
            data[i], data[pivot_index] = data[pivot_index], data[i]

    data[left], data[pivot_index] = data[pivot_index], data[left] 
    return pivot_index

def QuickSort(data, left, right, reverse=False):
    if right <= left:
        return

    # randomize the left pivot value
    pivot = left + random.randint(0, right - left)
    data[left], data[pivot] = data[pivot], data[left]

    pi = QuickSort_Partition2(data, left, right, reverse)
    QuickSort(data, left, pi - 1, reverse)
    QuickSort(data, pi + 1, right, reverse)
```

```python
# merge two sorted array and build one sorted array
def MergeArray(data, left, mid, right, reverse=False):
    leftsize = mid - left + 1
    rightsize = right - mid
    leftarr = data[left : mid + 1]
    rightarr = data[mid + 1 : right + 1]
    
    compareFunc = lambda a, b : a < b
    if reverse: compareFunc = lambda a, b : a > b

    # copy the smallest item from leftarr and rightarr
    k = left
    i = j = 0
    while i < leftsize and j < rightsize:
        if compareFunc(leftarr[i], rightarr[j]):
            data[k] = leftarr[i]
            i += 1
        else:
            data[k] = rightarr[j]
            j += 1
        k += 1

    # copy remaining leftarr element
    while i < leftsize:
        data[k] = leftarr[i]
        i += 1
        k += 1

    # copy remaining rightarr element
    while j < rightsize:
        data[k] = rightarr[j]
        j += 1
        k += 1

def MergeSort(data, left, right, reverse=False):
    if right <= left:
        return

    mid = left + int((right - left) / 2)
    MergeSort(data, left, mid, reverse)
    MergeSort(data, mid + 1, right, reverse)
    MergeArray(data, left, mid, right, reverse)
```

```python
def Heapify(data, parent, length, minheap=False):
    left = (parent * 2) + 1
    right = (parent * 2) + 2
    swapChild = parent
    
    compareFunc = lambda a, b : a > b
    if minheap: compareFunc = lambda a, b : a < b
    
    if left < length and compareFunc(data[left], data[swapChild]):
        swapChild = left

    if right < length and compareFunc(data[right], data[swapChild]):
        swapChild = right

    if swapChild != parent:
        # swap parent with largest/smallest child
        data[parent], data[swapChild] = data[swapChild], data[parent]

        # heapify again from the swapped node
        Heapify(data, swapChild, length, minheap)

def HeapSort(data, reverse=False):
    # create Max, Min heap starting from bottom internal node
    length = len(data)
    latParent = int(length / 2) - 1
    
    while latParent >= 0:
        Heapify(data, latParent, length, reverse);
        latParent -= 1
    
    lastChild = length - 1
    while lastChild > 0:
        # swap the max value (root node)
        # with the last unsorted position
        data[0], data[lastChild] = data[lastChild], data[0]

        # create max heap again from root 
        # to remaining unsorted items 
        Heapify(data, 0, lastChild, reverse)
        lastChild -= 1
```

```python
numbers = [5, 3, 1, 4, 6, 9, 7, 0, 2, 8]
print('Inputs   :', numbers)

QuickSort(numbers, 0, len(numbers) - 1, False)
print('QuickSort:', numbers)

MergeSort(numbers, 0, len(numbers) - 1, True)
print('MergeSort:', numbers)

HeapSort(numbers, False)
print('HeapSort :', numbers)
```
